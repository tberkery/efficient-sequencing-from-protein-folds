# Compress Protein Folding Sequence to non-repeating letters

```python
def compress_sequence(fileName):
    with open(fileName, 'r') as ifile:
        seq = ifile.read()
    
    shortSeq = seq[0]
    runLengths = [1]
    j = 0 # index in shortSeq
    
    for i in range(1, len(seq)):
        if seq[i] == seq[i - 1]:
            runLengths[j] += 1
        else:
            shortSeq = shortSeq + seq[i]
            runLengths.append(1)
            j += 1
    
    return shortSeq, runLengths
            
    
sequence, runLen = compress_sequence("sequence_20231111.txt")
print("sequence length:", len(sequence)) 
print(sequence)
```
